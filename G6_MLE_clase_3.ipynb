```python
import pandas as pd
from pycaret.regression import *
```

```python
#1. Cargamos la data:
df = pd.read_csv('precio_departamentos.csv')
```

```python
df.head()
```

```python
#Inicializar el experimiento
reg  = setup(data=df,target='precio',session_id=123,normalize=True,categorical_features=['distrito'])
```

```python
#3. Comparar modelos automaticamente:
best_model = compare_models()
```

```python
lr = create_model('lr')
```

```python
#5.Ajustar el modelo
tuned_lr = tune_model(lr)
```

```python
#6. Evaluar visualmente
evaluate_model(tuned_lr)
```

```python
#prediccin:
preds= predict_model(tuned_lr)
print(preds.head())
```

```python
#1. En promedio, el modelo se equivoca en un 5.76% del precio real
#2. El modelo explica el 97.43% de la varianza del precio
#3. Error logaritmico (0.0807) es bajo; buen performance en escalas grandes.

#Conclusin: El modelo tiene muy buen ajuste global. Es explicativo (R2 alto) y preciso (MAPE pequeo)
```

```python
#8. Guardar el modelo 
save_model(tuned_lr,'modelo_precio_final')
```
